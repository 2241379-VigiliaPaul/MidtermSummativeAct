# Data Preprocessing and Splitting Notebook

This notebook demonstrates each step of preprocessing and data splitting for the Telco Customer Churn dataset.

## Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import os

print("Libraries imported successfully!")
```

## Step 2: Load Raw Data

```python
# Load the raw Telco Customer Churn dataset
from pathlib import Path

# Get the notebook directory and project root
notebook_dir = Path.cwd()
project_root = notebook_dir.parent

raw_data_path = str(project_root / 'data' / 'raw' / 'Raw_Telco_Customer.csv')
df = pd.read_csv(raw_data_path)

print(f"Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

## Step 3: Display Raw Data Columns

```python
print("Columns in raw dataset:")
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")
```

## Step 4: Display First 5 Rows of Raw Data (Screenshot this)

```python
df.head()
```

## Step 5: Display Raw Data Info (Screenshot this)

```python
print("Raw Data Info:")
df.info()
```

## Step 6: Check Missing Values in Raw Data

```python
print("Missing values per column:")
df.isnull().sum()
```

## Step 7: Drop customerID Column

```python
# Drop customerID as it's not needed for analysis
df = df.drop(columns=['customerID'])
print("customerID column dropped")
print(f"New shape: {df.shape}")
```

## Step 8: Convert TotalCharges to Numeric

```python
# Convert TotalCharges to numeric, fill NaN with 0
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)

print("TotalCharges converted to numeric")
print(f"Data type: {df['TotalCharges'].dtype}")
print(f"Missing values filled: {(df['TotalCharges'] == 0).sum()}")
```

## Step 9: Encode gender (Female=0, Male=1)

```python
df['gender'] = df['gender'].map({'Female': 0, 'Male': 1})
print("gender encoded: Female=0, Male=1")
print(f"Value counts:")
print(df['gender'].value_counts().sort_index())
```

## Step 10: Encode Partner (No=0, Yes=1)

```python
df['Partner'] = df['Partner'].map({'No': 0, 'Yes': 1})
print("Partner encoded: No=0, Yes=1")
print(f"Value counts:")
print(df['Partner'].value_counts().sort_index())
```

## Step 11: Encode Dependents (No=0, Yes=1)

```python
df['Dependents'] = df['Dependents'].map({'No': 0, 'Yes': 1})
print("Dependents encoded: No=0, Yes=1")
print(f"Value counts:")
print(df['Dependents'].value_counts().sort_index())
```

## Step 12: Encode PhoneService (No=0, Yes=1)

```python
df['PhoneService'] = df['PhoneService'].map({'No': 0, 'Yes': 1})
print("PhoneService encoded: No=0, Yes=1")
print(f"Value counts:")
print(df['PhoneService'].value_counts().sort_index())
```

## Step 13: Encode MultipleLines (No=0, Yes=1, No phone service=2)

```python
df['MultipleLines'] = df['MultipleLines'].map({'No': 0, 'Yes': 1, 'No phone service': 2})
print("MultipleLines encoded: No=0, Yes=1, No phone service=2")
print(f"Value counts:")
print(df['MultipleLines'].value_counts().sort_index())
```

## Step 14: Encode InternetService (No=0, DSL=1, Fiber optic=2)

```python
df['InternetService'] = df['InternetService'].map({'No': 0, 'DSL': 1, 'Fiber optic': 2})
print("InternetService encoded: No=0, DSL=1, Fiber optic=2")
print(f"Value counts:")
print(df['InternetService'].value_counts().sort_index())
```

## Step 15: Encode Service Columns (Conditional)

```python
# Encode service columns based on InternetService
# If InternetService is 0 (No), service value = 2
# Otherwise, Yes = 1, No = 0

service_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in service_cols:
    df[col] = df.apply(
        lambda row: 2 if row['InternetService'] == 0 else (1 if row[col] == 'Yes' else 0), 
        axis=1
    )
    print(f"{col} encoded")

print(f"\nService columns encoding logic:")
print("No internet service -> 2")
print("Yes -> 1")
print("No -> 0")
```

## Step 16: Show Service Column Value Counts (Screenshot this)

```python
for col in ['OnlineSecurity', 'OnlineBackup', 'TechSupport']:
    print(f"\n{col}:")
    print(df[col].value_counts().sort_index())
```

## Step 17: Encode Contract (One year=1, Two year=2, Month-to-month=3)

```python
df['Contract'] = df['Contract'].map({'One year': 1, 'Two year': 2, 'Month-to-month': 3})
print("Contract encoded: One year=1, Two year=2, Month-to-month=3")
print(f"Value counts:")
print(df['Contract'].value_counts().sort_index())
```

## Step 18: Encode PaperlessBilling (No=0, Yes=1)

```python
df['PaperlessBilling'] = df['PaperlessBilling'].map({'No': 0, 'Yes': 1})
print("PaperlessBilling encoded: No=0, Yes=1")
print(f"Value counts:")
print(df['PaperlessBilling'].value_counts().sort_index())
```

## Step 19: Encode PaymentMethod (Electronic check=0, Mailed check=1, Bank transfer=2, Credit card=3)

```python
df['PaymentMethod'] = df['PaymentMethod'].map({
    'Electronic check': 0, 
    'Mailed check': 1, 
    'Bank transfer (automatic)': 2, 
    'Credit card (automatic)': 3
})
print("PaymentMethod encoded:")
print("Electronic check -> 0")
print("Mailed check -> 1")
print("Bank transfer (automatic) -> 2")
print("Credit card (automatic) -> 3")
print(f"\nValue counts:")
print(df['PaymentMethod'].value_counts().sort_index())
```

## Step 20: Encode Target Variable - Churn (No=0, Yes=1)

```python
df['Churn'] = df['Churn'].map({'No': 0, 'Yes': 1})
print("Churn encoded: No=0, Yes=1")
print(f"Value counts:")
print(df['Churn'].value_counts())
print(f"\nChurn rate: {df['Churn'].mean():.2%}")
```

## Step 21: Rename Columns with Descriptive Names

```python
# Rename columns to match cleaned dataset format
df = df.rename(columns={
    'gender': 'gender( m = 1, f = 0)',
    'SeniorCitizen': 'SeniorCitizen (Y=1, N=0)',
    'Partner': 'Partner (Y=1, N=0)',
    'Dependents': 'Dependents (Y=1, N=0)',
    'PhoneService': 'PhoneService (Y=1, N=0)',
    'MultipleLines': 'MultipleLines (Y=1, N=0 , NS = 2)',
    'InternetService': 'InternetService ( N = 0, DSL = 1, FBR OPT. = 2',
    'OnlineSecurity': 'OnlineSecurity ( 0 internet serv. =2)',
    'OnlineBackup': 'OnlineBackup ( 0 internet serv. =2)',
    'DeviceProtection': 'DeviceProtection ( 0 internet serv. =2)',
    'TechSupport': 'TechSupport ( 0 internet serv. =2)',
    'StreamingTV': 'StreamingTV ( 0 internet serv. =2)',
    'StreamingMovies': 'StreamingMovies ( 0 internet serv. =2)',
    'Contract': 'Contract ( 1yr = 1, 2 yr = 2 , month to month = 3',
    'PaperlessBilling': 'PaperlessBilling ( Y = 1, N = 0)',
    'PaymentMethod': 'PaymentMethod (eCheck = 0, mCheck = 1, bTransfer =2, card = 3)'
})

print("Columns renamed successfully!")
print(f"Total columns: {len(df.columns)}")
```

## Step 26: Save Cleaned Data

```python
# Create processed directory if needed
processed_dir = str(project_root / 'data' / 'processed')
os.makedirs(processed_dir, exist_ok=True)

# Save cleaned data
cleaned_data_path = str(project_root / 'data' / 'processed' / 'Cleaned_Telco_Customer.csv')
df.to_csv(cleaned_data_path, index=False)

print(f"Cleaned data saved to: {cleaned_data_path}")
print(f"Shape: {df.shape}")
```

## Step 27: First Data Split - 90% Train/Test, 10% Validation

```python
# Split data according to requirements:
# - 90% for training/testing (will be split further)
# - 10% unseen validation

target_column = 'Churn'

# First split: 6339 train_test, 704 validation (stratified)
train_test, validation = train_test_split(
    df, test_size=704, random_state=42, stratify=df[target_column]
)

print("First split completed:")
print(f"Train/Test pool (90%): {len(train_test)} samples")
print(f"Validation (10%): {len(validation)} samples")
print(f"Total: {len(train_test) + len(validation)} samples")
```

## Step 28: Validation Set Churn Distribution (Screenshot this)

```python
print("Validation set churn distribution:")
print(validation[target_column].value_counts())
print(f"\nChurn rate: {validation[target_column].mean():.2%}")
```

## Step 29: Second Data Split - 80% Train, 20% Test (from the 90% pool)

```python
# Second split: from the 90%, create 80% train and 20% test
# 5071 train, 1268 test
train, test = train_test_split(
    train_test, test_size=1268, random_state=42, stratify=train_test[target_column]
)

print("Second split completed:")
print(f"Train (80% of 90%): {len(train)} samples")
print(f"Test (20% of 90%): {len(test)} samples")
print(f"Subtotal: {len(train) + len(test)} samples")
```

## Step 30: Test Set Churn Distribution (Screenshot this)

```python
print("Test set churn distribution:")
print(test[target_column].value_counts())
print(f"\nChurn rate: {test[target_column].mean():.2%}")
```

## Step 31: Training Set Churn Distribution (Screenshot this)

```python
print("Training set churn distribution:")
print(train[target_column].value_counts())
print(f"\nChurn rate: {train[target_column].mean():.2%}")
```

## Step 32: Final Split Summary (Screenshot this)

```python
# Display final split summary
print("=" * 50)
print("FINAL DATA SPLIT SUMMARY")
print("=" * 50)
print(f"Train:      {len(train)} samples ({len(train)/len(df)*100:.1f}%)")
print(f"Test:       {len(test)} samples ({len(test)/len(df)*100:.1f}%)")
print(f"Validation: {len(validation)} samples ({len(validation)/len(df)*100:.1f}%)")
print("-" * 50)
print(f"Total:      {len(train) + len(test) + len(validation)} samples")
print("=" * 50)
print("\nSplit purpose:")
print("- Train (5071): Used for model training")
print("- Test (1268): Used for model evaluation")
print("- Validation (704): Unseen data for final validation")
```

## Step 33: Save Data Splits to CSV Files

```python
# Create splits directory
splits_dir = str(project_root / 'data' / 'splits')
os.makedirs(splits_dir, exist_ok=True)

# Save all splits
train.to_csv(str(project_root / 'data' / 'splits' / 'train.csv'), index=False)
test.to_csv(str(project_root / 'data' / 'splits' / 'test.csv'), index=False)
validation.to_csv(str(project_root / 'data' / 'splits' / 'validation.csv'), index=False)

print("Data splits saved successfully!")
print("\nFiles created:")
print("- data/splits/train.csv")
print("- data/splits/test.csv")
print("- data/splits/validation.csv")
```
